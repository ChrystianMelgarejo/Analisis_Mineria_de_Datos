<h1 style="text-align: left;">Dashboard informativo sobre productos reportados a DOE. </h1>
<h2 style="text-align: left;">Periodo Enero - Junio 2020.</h2>

<div class="alert alert-block alert-info">
    <h2>Introducción:</h2>    
    <p>El Directorio de rendimiento de productos certificados de AHRI (www.ahridirectory.org) es la fuente confiable de equipos HVACR con rendimiento certificado.

En el lado público del Directorio, los ingenieros, contratistas, reguladores y consumidores pueden buscar por líneas de productos, calificaciones certificadas o marcas para encontrar la información que necesitan, cuando la necesitan. En el lado autenticado, los fabricantes cuentan con herramientas para cargar, administrar y ver sus datos certificados por AHRI (incluidos los detalles a los que no puede acceder el público).
    <h2>Departamento de Energía de los Estados Unidos:</h2>    
    AHRI se encarga de reportar de manera diaria todos los productos certificados que califican dentro de los programas definidos por DOE. En la actualidad existen 42 programas de los cuales reconoceremos en el ejercicio como <strong>Report Process Id 01</strong> al <strong>Report Process Id 42</strong>.
    
El propósito del análisis es determinar que programa de los 42 existentes fue el más demandante para AHRI durante el periodo  Enero - Junio del 2020.</p>
</div>

<h2>Dashboard 01:</h2>

<div class="alert alert-block alert-warning">
<b>Detalles:</b><br> 
Se agrupó la información por el campo "ReportProcessId" y se calculó el total de productos certificados durante el periodo seleccionado.
</div>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_SourceData = pd.read_csv("DOEdata.csv") #Import data from CSV file.
df_SourceData["SubmissionDate"] = pd.to_datetime(df_SourceData["SubmissionDate"]).dt.date # Redefine data type

minReportProcessId = min(df_SourceData["ReportProcessId"])
maxReportProcessId = max(df_SourceData["ReportProcessId"])

#Configuration of the graphic
plt.figure(figsize=(9,6))
plt.title("DOE Information", fontsize=20)
plt.xlabel("Report Process Id")
plt.ylabel("Number of Models")
plt.xticks(np.arange(minReportProcessId, maxReportProcessId, 1.0))
plt.xticks(fontsize=8, rotation=90)

plt.bar(df_SourceData["ReportProcessId"],df_SourceData["TotalModels"])
plt.show()
```

<div class="alert alert-block alert-success">
<b>Conclusión:</b><br> 
Se puede observar que el "ReportProcessId" = 1 es el que más información reportó durante el periodo seleccionado.
</div>

<h2>Dashboard 02:</h2>

<div class="alert alert-block alert-warning">
<b>Detalles:</b><br> 
Se agrupó la información en dos bloques. 
    <br/>1) Todos los productos reportados dentro del ReportProcessId = 1 
    <br/>2) El resto de los productos pertenecientes a los otros programas.
</div>

```python
reportProcessId1_Totals = df_SourceData["TotalModels"].loc[df_SourceData["ReportProcessId"] == 1].sum()
allOtherPrograms_Totals = df_SourceData["TotalModels"].loc[df_SourceData["ReportProcessId"] != 1].sum()

#Configuration of the graphic
plt.figure(figsize=(9,6))
plt.title("DOE Information", fontsize=20)

labels = ["Report Process Id = 1","All other programs"]
colors = ["#ff9999","#66b3ff"]
explode = (0, 0.1)

#Link variables to the plot and then display.
plt.pie([reportProcessId1_Totals, allOtherPrograms_Totals], explode = explode,labels = labels, colors = colors, autopct="%.2f %%", textprops={'color':"black", 'size':"14"}, shadow=True, startangle=90)
plt.show()
```

<div class="alert alert-block alert-success">
<b>Conclusión:</b><br> 
Se puede observar que aproximadamente 1/3 de los productos reportados a DOE pertenecen al programa identificado como ReportProcessId = 1.
</div>

<h2>Dashboard 03:</h2>

<div class="alert alert-block alert-warning">
<b>Detalles:</b><br> 
Se agrupó la información por el campo "ReportProcessId" = 1. El propósito es definir en que mes se tuvo el mayor número de modelos reportados a DOE.
</div>

```python
filter = df_SourceData["ReportProcessId"] == 1
df_DataByReportProcess = df_SourceData.where(filter) #Create a new subset of data specifically for the ReportProcessId = 1 

df_RP01_TotalsPerMonth = pd.DataFrame(columns = ["TotalModels", "SubmissionDate"]) #Create empty DataFrame
df_RP01_TotalsPerMonth["TotalModels"] = df_DataByReportProcess["TotalModels"] #Populate column
df_RP01_TotalsPerMonth["SubmissionDate"] = pd.DatetimeIndex(df_DataByReportProcess['SubmissionDate']).month #Populate column

df_RP01_TotalsPerMonth = df_RP01_TotalsPerMonth.groupby(['SubmissionDate']).sum()

#Configuration of the graphic
plt.figure(figsize=(9,6))
plt.title("DOE Information", fontsize=20)
plt.xlabel("Month of the Year")
plt.ylabel("Number of Models")

labels = ["Jan","Feb","Mar","Apr","May", "Jun"]

#Link dataframe to the plot and then display.
plt.plot(df_RP01_TotalsPerMonth.index, df_RP01_TotalsPerMonth["TotalModels"], 'g.-', label="Report Process Id 1")
plt.xticks(df_RP01_TotalsPerMonth.index, labels)
plt.legend()
plt.show()
```

<div class="alert alert-block alert-success">
<b>Conclusión:</b><br> 
Se puede observar que en el mes de enero se transmitieron mayor número de modelos.
</div>


<h2>Dashboard 04:</h2> 

<div class="alert alert-block alert-warning">
<b>Detalles:</b><br> 
Se agrupó la información por el campo "ReportProcessId" = 1 y por el mes de "enero". El propósito es definir en que día de la semana se tuvo el mayor número de modelos reportados a DOE.
</div>

```python
filter = (df_SourceData["ReportProcessId"] == 1) & (df_SourceData["SubmissionDate"] >= pd.to_datetime("2020-01-01").date()) & (df_SourceData["SubmissionDate"] <= pd.to_datetime("2020-01-31").date())
df_DataByReportProcessAndMonth = df_SourceData.where(filter) #Create a new subset of data

df_RP01_TotalsPerDayOfWeek = pd.DataFrame(columns = ["TotalModels", "SubmissionDate"]) #Create empty DataFrame
df_RP01_TotalsPerDayOfWeek["TotalModels"] = df_DataByReportProcessAndMonth["TotalModels"] #Populate column
df_RP01_TotalsPerDayOfWeek["SubmissionDate"] = pd.DatetimeIndex(df_DataByReportProcessAndMonth['SubmissionDate']).weekday #Populate column

df_RP01_TotalsPerDayOfWeek = df_RP01_TotalsPerDayOfWeek.groupby(['SubmissionDate']).sum()

#Configuration of the graphic
plt.figure(figsize=(9,6))
plt.title("DOE Information", fontsize=20)
plt.xlabel("Day of the Week")
plt.ylabel("Number of Models")
color = ["#2573e8","#0d2bbf","#450dbf","#600dbf","#ad0dbf","#bf0d66"]
tick_label=["M","T","W","Th","F","S",]

#Link dataframe to the plot and then display.
plt.bar(df_RP01_TotalsPerDayOfWeek.index, df_RP01_TotalsPerDayOfWeek["TotalModels"], color=color, tick_label=tick_label)
plt.show()
```

<div class="alert alert-block alert-success">
<b>Conclusión:</b><br> 
Se puede observar que en los días miercoles del mes de enero se reportaron más modelos a DOE.
</div>
